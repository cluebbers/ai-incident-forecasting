```python
import numpy as np
import matplotlib.pyplot as plt


def plot_ytd_assimilation(ax, observed_months=10, last_observed_month=None):
    """Plot a YTD assimilation diagram on the provided axes."""
    months = [ "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
    # Simple illustrative values to show accumulation through the year
    values = np.array([ 8, 9, 10, 11, 12, 13, 14], dtype=float)

    if last_observed_month is not None:
        last_idx = max(0, min(11, int(last_observed_month) - 1))
    else:
        last_idx = max(0, min(11, int(observed_months) - 1))

    x = np.arange(len(months))
    actual_mask = x <= last_idx
    fill_mask = x > last_idx

    actual_color = "#1f4e79"
    fill_color = "#e07a2a"

    ax.bar(x[actual_mask], values[actual_mask], color=actual_color, edgecolor=actual_color)
    ax.bar(
        x[fill_mask],
        values[fill_mask],
        color=fill_color,
        edgecolor=fill_color,
        hatch="///",
        alpha=0.45,
        linewidth=1.0,
    )

    ax.set_xticks(x)
    ax.set_xticklabels(months, fontsize=8)
    ax.set_yticks([])
    ax.set_xlim(-0.6, len(months) - 0.4)
    ax.set_ylim(0, values.max() * 1.12)

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["bottom"].set_color("#4a4a4a")
    ax.tick_params(axis="x", length=0, pad=6)

    ax.text(0.02, 0.96, "YTD observed", transform=ax.transAxes, ha="left", va="top", fontsize=8, color=actual_color)
    ax.text(0.58, 0.96, "Filled", transform=ax.transAxes, ha="left", va="top", fontsize=8, color=fill_color)
```

```python
# Example usage for a vertical 4:3 subplot
fig, ax = plt.subplots(figsize=(3, 4))
plot_ytd_assimilation(ax, observed_months=6)
fig.tight_layout()
plt.show()
```

```python
# Example usage for a vertical 4:3 subplot
fig, ax = plt.subplots(figsize=(3, 4))
plot_ytd_assimilation(ax, last_observed_month=10)
fig.tight_layout()
plt.show()
```
